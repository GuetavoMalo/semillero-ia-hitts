**Definiciones**

- SSE (Sum of Squared Errors): SSE = Σ (y_i - ŷ_i)^2 . Variación no explicada por el modelo; mide el tamaño de los residuos.
Captura de pantalla 2025-10-08 a la(s) 11.49.18.png

- SSR (Sum of Squares due to Regression): SSR = Σ (Ŷ_i - ȳ)^2 . Variación explicada por el modelo respecto a la media.
Captura de pantalla 2025-10-08 a la(s) 11.56.29.png


- Syy (Total Sum of Squares): Syy = Σ (y_i - ȳ)^2 . Variabilidad total de la variable respuesta.
Captura de pantalla 2025-10-08 a la(s) 11.57.56.png



- Identidad clave: Syy = SSR + SSE (cuando el modelo incluye intercepto).
Por qué se usan para evaluar regresión
Captura de pantalla 2025-10-08 a la(s) 11.59.26.png


- Separan la variabilidad total en “explicada” ( SSR ) y “no explicada” ( SSE ), permitiendo medir qué tanto aporta el modelo.

- Permiten construir métricas estándar de calidad del ajuste (como R^2 , MSE ) y pruebas de significancia (estadístico F ).

- Facilitan comparar modelos: menor SSE y mayor SSR suele indicar mejor capacidad explicativa, manteniendo ojo en complejidad.

Relaciones y métricas derivadas

- Coeficiente de determinación: R^2 = SSR / Syy = 1 - SSE / Syy . Proporción de variabilidad explicada por el modelo.

- Error medio cuadrático residual: MSE = SSE / (n - p - 1) , con n observaciones y p predictores (sin contar el intercepto).

- Media de cuadrados debido a la regresión: MSR = SSR / p .
- Prueba F del modelo global: F = (MSR) / (MSE) = (SSR/p) / (SSE/(n - p - 1)) . Evalúa si al menos un predictor tiene efecto.

- R^2 ajustado (para penalizar complejidad): 1 - [(SSE/(n - p - 1)) / (Syy/(n - 1))] .

Interpretaciones prácticas

- SSE pequeño: residuos más cercanos a cero; mejor ajuste (pero cuidado con sobreajuste).
- SSR grande: el modelo captura más estructura de los datos respecto a la media.
- Syy es fijo para un conjunto de datos; sirve como referencia para cuantificar la fracción explicada vs. no explicada.
- Si el modelo no incluye intercepto, la descomposición Syy = SSR + SSE puede no sostenerse; suele ser recomendable incluirlo.
- A medida que añades predictores, SSR tiende a subir y SSE a bajar; usa R^2 ajustado y F para validar mejoras reales.
